```python
import seaborn as sns
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import missingno as msno
from numpy import nansum
from numpy import nanmean
```

* Load all CSV files into DataFrames

```python
all_DataframesFile = pd.read_csv("all_DataframesFile.csv")
all_DataframesFileByYear = pd.read_csv("all_DataframesFileByYear.csv")
all_DataframesFileByMonth = pd.read_csv("all_DataframesFileByMonth.csv")
```

```python
yearStart=1980
yearEnd=2021

all_DataframesFileByYear[(all_DataframesFileByYear['year']>=yearStart)&(all_DataframesFileByYear['year']<=yearEnd)].plot(
    x='year',
    y=["Kinneret_Level_end_of_year"],
    figsize=(40, 5),
    title='Kineret Levels',
    grid=True
)
plt.axhline(y=-208.80, color='r', linestyle='-')
plt.axhline(y=-212.00, color='r', linestyle='-')

all_DataframesFileByYear[(all_DataframesFileByYear['year']>=yearStart)&(all_DataframesFileByYear['year']<=yearEnd)].plot(
    x='year',
    y=['tmp_air_max'],
    title='Average maximum temperature value relative to years',
    figsize=(40, 5),
    grid=True
)

all_DataframesFileByYear[(all_DataframesFileByYear['year']>=yearStart)&(all_DataframesFileByYear['year']<=yearEnd)].plot(
    kind='bar',
    x='year',
    y=['Change_from_last_year'],
    title='Change from last year',
    figsize=(40, 5),
    grid=True
)
all_DataframesFileByYear[(all_DataframesFileByYear['year']>=yearStart)&(all_DataframesFileByYear['year']<=yearEnd)].plot(
    kind='bar',
    x='year',
    y=['rain_mm'],
    title='The amount of precipitation',
    figsize=(40, 5),
    grid=True
)
```

```python
#Multiplotting
plt.figure(1,figsize=[14,12])

# Figure num 1
plt.subplot(2,2,1)
sns.boxplot(data=all_DataframesFileByMonth,y="Change_from_last_month",x="month")
plt.title("      ")
plt.xlabel(" ")
plt.ylabel("  ")
plt.ylim([-1.5,1.5])

#figure num 2
plt.subplot(2,2,2)
sns.boxplot(data=all_DataframesFileByMonth,y="sns_drt",x="month")
plt.title("      ")
plt.xlabel(" ")
plt.ylabel("   ")
```
